### Create the function to search according the title and artist 

```python
def get_song_id (song_name, artist_name):
    try: 
        results = sp.search(q='tracks:'+song_name+'artist:'+artist_name, limit =1)
        return results ['tracks']['items'][0]['id']
    except:
        return "None"
```

```python
def search_song_complex():
    
    # prompt the user to enter the song and artist names
    title = input('Enter song name:')
    artist = input('Enter artist name:')
    
    #search for the song using the title and artist
    results = sp.search(q='track:"{}" artist:"{}"'.format(title, artist), type='track', limit=10)
    
    #If there are multipe search results, let the user choose which one to use
    
    if len(results['tracks']['items'])>1:
        
        print('Multiple matches found. Please choose one:')
        
        for i, item in enumerate(results['tracks']['items']):
            print(i, item['name'], "-",item['artists'][0]['name'])
            
        selection = int(input("Enter selections:"))
        track_id = results['tracks']['items'][selection]['id']
        return 'spotify: track:' + track_id
        
    #If there are multiple search results, let the user choose which one to use
    elif len(results['tracks']['items']) == 1:
            track_id = results['tracks']['items'][0]['id']
            return 'spotify:track:' + track_id
    else:
        print("No match found")
        return ""
    
track_features = search_song_complex()
print(track_features)
    
```

### Create a function get_audio_features(list_of_songs) to obtain the audio features of a given list of songs (the content of list_of_songs can be the href/id/uri).

Create chunks of 50 songs

```python
title_not= []

for x in not_hot_songs ['title']:
    title_not.append(x)
    
chunk_size = 50

title_not_chunked = [title_not[i:i + chunk_size] for i in range(0,len(title_not), chunk_size)]
print(title_not_chunked)
len(title_not_chunked[1])
```

```python
title_top = []

for x in top_hot_songs['title']:
    title_top.append(x)

    
title_top_chunked = [title_top[i:i + chunk_size] for i in range (0, len(title_top), chunk_size)]
print(title_top_chunked)
len(title_top_chunked[1])
```

```python
artist_top = []

for x in top_hot_songs['artist']:
    artist_top.append(x)

    
artist_top_chunked = [artist_top[i:i + chunk_size] for i in range (0, len(artist_top), chunk_size)]
print(artist_top_chunked)
len(artist_top_chunked[1])
```

```python
artist_not = []

for x in not_hot_songs['artist']:
    artist_not.append(x)

    
artist_not_chunked = [artist_not[i:i + chunk_size] for i in range (0, len(artist_not), chunk_size)]
print(artist_not_chunked)
len(artist_not_chunked[1])
```

```python
def add_audio_features(title_chunked, artist_chunked):
    list_of_ids = []
    all_my_dict = {}
    for tit, art in zip(title_chunked, artist_chunked):    
        for t, a in zip(tit, art):
            new_id = get_song_id(t, a)
            #list_of_ids.append(new_id)
            #print(new_id)
            if new_id != "None":
                try:
                    my_dict = sp.audio_features(new_id)[0]
                    all_my_dict[new_id] = my_dict
                    all_my_dict[new_id]['song_name'] = t  # add song name to the dictionary
                    all_my_dict[new_id]['artist_name'] = a  # add artist name to the dictionary
                except:
                    print ("Couldn't get the audio features")
            
            else:
                print("Invalid URI ({}, {})".format(a, t))
            
        print('I will sleep for 10 seconds')
        sleep(10)
        
    df = pd.DataFrame(all_my_dict).T
    return df
```

```python
hot_songs_audio = add_audio_features(title_top_chunked, artist_top_chunked)
```
